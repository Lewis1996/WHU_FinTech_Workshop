## 0.2 股票分割

##### 股票分割的定义
按照已有股票数量的比例，免费向股东发行额外数量的股票从基本面的角度来看，股票分割似乎对公司的经济影响并不大。

##### 为什么进行股票分割？（市场有效假设下）
1. 信息传递假设认为管理层利用股票分割来传达有关公司积极前景的私人信息
2. 流动性假设则认为管理层利用股票分割来提高公司股票的流动性。

##### 市场对股票分割信息反应的期望时间段
1. 法马等人（1969年）将“月份0”定义为分割生效日期所在的月份。
2. 法马等人（1969年）提到，对于他们样本中的“随机抽取的52次分割”，“分割宣布日期与生效日期之间的中位时间为44.5天”（FFJR的分析使用月度数据，因此分割生效日期前44.5天的中位宣布日期暗示中位分割是在月份-2中宣布的）
3. 如果股票分割具有信息内容，我们预计有效市场会在分割宣布时做出反应，而不是在生效时。


##### 讨论问题
考虑以下的替代理论：
1. 理论 A：公司喜欢保持其股价在一定范围内。当股价上涨超过某一阈值时，公司可能会启动股票分割，并提前数周宣布生效日期。公司不使用股票分割来传递有关公司前景的私人信息。
2. 理论 B：公司利用股票分割来传递有关公司前景的私人信息。公司将在生效日期前数周宣布分割。
3. 理论 C：资本市场参与者未完全调整股票分割的影响，在某种程度上倾向于锚定在股票分割前的价格上。公司不使用股票分割来传递有关公司前景的私人信息。（基于这种假设，公司可能会提前公布股票分割，以给市场更多的时间来适应和理解这一决定。通过提前公布股票分割，公司可以向市场传达他们的意图，并提供更多的信息，帮助市场更好地理解分割的动机和可能的影响。这样做也可以减少市场对分割公布时的突然性反应，有助于平稳过渡。此外，提前公布还可以给股东和投资者更多时间来调整他们的投资策略和预期，从而减少分割公布后的市场波动。）

## 0.3 股息政策
##### 股息增加和削减传递的信息
1. Black（1976年）总结了对股息的常见理解：“出于某种原因，经理和董事不愿意削减股息。因此，他们只会在认为公司的前景足够好以支持较长时间的较高股息时才会提高股息。只有当他们认为快速恢复的前景不佳时，他们才会削减股息”。经理和董事拥有内幕信息，因此，如果他们按照这一理论行事，股息政策将传达部分这些信息
2. 法马等人（1969年）支持了对股息的这一解释：“研究表明，一旦股息增加，大型公司极不愿意减少它们，除非处于极端情况下。董事们似乎通过只在他们确信能够将来维持它们的能力非常强时才增加股息来对冲这种股息削减的可能性，即只有在他们强烈认为未来的收益足以维持股息以其新的更高水平时才增加股息。因此，股息变化可以假定为向市场传达了有关管理层对公司长期收益和股息支付潜力的重要信息。

##### 股票分割和股息之间的关系
法马等人（1969年）表示，“过去大部分股票分割都紧随着股息增加。”法马等人（1969年）提供了以下解释：“当股票分割被宣布或预期时，市场会将其解释为（而且确实如此），大大提高了股息很快将被大幅增加的可能性。（实际上，在许多情况下，分割和股息增加将同时宣布。）”

### 2.1 股票分割样本

##### 获取股票分割信息
法马等人（1969年）将“股票分割”定义为一种股票交换，其中每四股票原先的股份至少分配五股。因此，这种对于分割的定义包括所有股票分红比例为25%或更高的情况。

```python
## distcd：选择特定的股息和股票分红
splits = (dsedist[(dsedist['exdt'] >= '1927-01-01') & 
                  (dsedist['exdt'] <= '1959-12-31') & 
                  (dsedist['distcd'].isin([5523, 5533])) & 
                  (dsedist['facshr'] >= 0.25)]
          .groupby(['permno','exdt'],group_keys=False)
          .agg({'facshr': 'sum'})
          .reset_index())
splits['split']=True
splits
```

1. 数据仅涵盖纽约证券交易所上市的普通股（ 纽约证券交易所上市的股票具有 exchcd == 1，普通股则是指 shrcd 的第一个字符为 1）
2. 规定要求，为了符合测试的条件，分割证券必须在分割前后至少在交易所上市12个月

```python
nyse_stocks = (stocknames[(stocknames['exchcd'] == 1) & 
                             (stocknames['shrcd'].astype(str).str[:1] == "1")]
               .get(['permno', 'namedt', 'nameenddt']))
nyse_stocks
```

```python
nyse_splits_raw = (splits
                   .merge(nyse_stocks,
                          how='inner',
                          on='permno'))

nyse_splits_raw = nyse_splits_raw[(nyse_splits_raw['exdt'] >= nyse_splits_raw['namedt']) & 
                                  (nyse_splits_raw['exdt'] <= nyse_splits_raw['nameenddt'])]

nyse_splits_raw
```

需要创建一个“月份”变量，以便我们可以对齐 crsp.msf（月度回报数据） 和 nyse_splits_raw 上的观察数据。


```python
# 创建了一个名为 month_indexes 的表，其中包括每个月的“索引”，该索引也将用于进行前后几个月的算术运算
msi['month'] = pd.to_datetime(msi['date'].dt.strftime('%Y-%m'))
msi = msi.sort_values(['month'])
msi['month_index']=range(1,len(msi)+1)
month_indexes = msi.get(['date','month','month_index'])
month_indexes
```

```python
nyse_splits_raw['month']= pd.to_datetime(nyse_splits_raw['exdt'].dt.strftime('%Y-%m'))

nyse_splits = (nyse_splits_raw
               .merge(month_indexes,
                      how='inner',
                      on='month')
               .rename(columns={'month_index':'ex_month_index'})
               .drop(['namedt','nameenddt','date'],axis=1))
nyse_splits
```

将 crsp.msf 限制为纽约证券交易所的股票，并且月度回报不缺失，并增加了变量 month_index

```python
nyse_msf = (msf
            .dropna(subset=['ret'])
            .merge(nyse_stocks,
                   how='inner',
                   on='permno')
            .query('date >= namedt & date <= nameenddt')
            .merge(month_indexes,
                   how='inner',
                   on='date')
            .get(['permno','month_index','date','ret']))
nyse_msf
```

将来自 nyse_splits 的分割数据与来自 nyse_msf 的回报数据进行合并。我们创建一个变量（month_rel_ex），用于衡量分割和回报之间的月份数

```python
split_return_data = (nyse_splits
                     .merge(nyse_msf,
                            how='left',
                            on='permno')
                     .assign(month_rel_ex=lambda x: x['month_index']-x['ex_month_index'])
                     .get(['permno','exdt','month_rel_ex','date','ret']))

split_return_data
```

## 2.2 估计超额回报
1. FFJR通过将每支股票的总回报的对数回归到市场指数的总回报的对数，并取残差来估计市场模型
2. Fama等人（1969年）使用的市场指数是费舍尔的综合投资绩效指数，该指数作为Fisher（1966年）的表A1提供。要使用这个表格，我们需要手工输入其中的数百个数字，因为它没有以可机读的形式提供；我们将使用CRSP在crsp.msi中提供的标准指数之一（vwretd和ewretd，我们将重点关注vwretd）

```python
index_returns = (msi.get(['date','vwretd','ewretd']))
index_returns
```

1. 在研究中，法马等人（1969年）提到了一种称为“规范误差”的问题，即非零超额回报可能导致分析模型的错误规范化。为了解决这个问题，他们决定在分析中排除特定时间段的数据。具体来说，他们排除了“对于所有证券，在分割前十五个月的数据，以及在分割后跟着股利减少的情况下，在分割后十五个月的数据”。
2. 这意味着他们排除了分割前十五个月的数据，但不排除分割后的任何月份的数据。这是因为在分割后的月份需要关于股利的数据，而这些数据在分割时是不可用的，直到稍后才能收集到。因此，排除这些月份可能会导致额外的前瞻性问题。

```python
omit_returns = (split_returns
                .assign(exclude = lambda x: ((x['month_rel_ex']>=-15) & (x['month_rel_ex']<=0)) )
                .groupby(['permno','date'],group_keys=False)
                .agg({'exclude':'any'})
                .reset_index())
omit_returns
```

```python
## 将 exclude 添加到 split_returns 数据中
## 因为（permno，date）组合可能会出现多次，但由于它涉及到多个分割，因此 month_rel_ex 的值可能不同
split_returns_reg = (split_returns
                     .merge(omit_returns,
                            how='inner',
                            on=['permno','date'])
                     .get(['permno','date','ret','exclude'])
                     .drop_duplicates())
split_returns_reg
## 还请注意，我们不会简单地从 exclude 为 TRUE 的数据集中删除行，因为我们希望计算这些月份的异常回报。
## 相反，我们仅通过在 subset = !exclude 中指定来排除这些观察结果，将这些观察结果从回归分析中排除
```

按照股票编号（permno）估计回归模型。与FFJR类似，我们通过将每支股票的总回报的对数回归到市场指数的总回报的对数中，并取残差来估计市场模型

```python
## python无法在回归内进行bool筛选
def split_reg_predict(data):
    reg_data = data[data['exclude']==False]
    reg_model= sm.OLS.from_formula(formula="lpr~1+lm", data=reg_data, missing='drop').fit()
    data['fit'] = reg_model
    data['predicted'] = reg_model.predict(data['lm'])
    data['resid'] = data['lpr'] - data['predicted']
    return data
```

```python
abnormal_returns = (split_returns_reg
                    .merge(index_returns,
                           how='left',
                           on='date')
                    .assign(lpr = lambda x: np.log(1+x['ret']),
                            lm = lambda x: np.log(1+x['vwretd']))
                    .get(['permno','date','lpr','lm','exclude'])
                    .groupby('permno',group_keys=False)
                    .apply(lambda x: x.assign(fit = split_reg_predict(x)['fit'],
                                              predicted = split_reg_predict(x)['predicted'],
                                              resid = split_reg_predict(x)['resid']))
                    .reset_index()
                    .get(['permno','date','resid']))
abnormal_returns 
## 与R结果相同
```

```python
## 将 abnormal_returns 与 nyse_splits 结合起来
## 过滤出距离相关分割超过30个月的观测值
table2_data = (abnormal_returns
               .merge(month_indexes,
                      how='inner',
                      on='date')
               .merge(nyse_splits,
                      how='left',
                      on='permno')
               .assign(month_gap = lambda x: x['month_index']-x['ex_month_index'])
               )
## month_gap表示残差数据是分割后n个月的数据，负值表示分割前
table2_data = table2_data[abs(table2_data['month_gap'])<=30].get(['permno','exdt','month_gap','resid']).reset_index(drop=True)

table2_data
```

```python
## 我们从 dsedist 中收集股利数据。普通股利与其他分配（例如清算股利或交换和重组）通过 distcd 的第一个数字进行区分，对于普通股利，该数字为1
## 对于给定的证券，在给定的月份可能会支付多个股利，因此我们按（permno，month）对股利进行聚合
div_months =(dsedist[dsedist['distcd'].astype(str).str[:1] == "1"]
             .assign(month = lambda x: pd.to_datetime(x['exdt'].dt.strftime('%Y-%m')))
             .groupby(['permno','month'],group_keys=False)            
             .agg({'divamt': lambda x: x.sum(skipna=True)})
             .reset_index())
div_months
## 月度股利数据
## 这个跟R有点不一样 R:920862
```

2. 在计算“每股未分割股份”股利时需要小心，因为如果一个公司在进行二比一分割前每个季度每股支付80美分，那么如果在分割后每个季度每股支付45美分（即每个原始股份90美分），我们认为它增加了股利

```python
## 筛选nyse股票
## cfacshr 调整股利
nyse_divs_raw = (msf
                 .merge(nyse_stocks,
                        how='inner',
                        on='permno')
                 .query('date >= namedt & date <= nameenddt')
                 .assign(month = lambda x: pd.to_datetime(x['date'].dt.strftime('%Y-%m')))
                 .get(['permno','date','month','cfacshr'])
                 .merge(div_months,
                        how='left',
                        on=['permno','month'])
                 .assign(divamt=lambda x: ((x['divamt'].fillna(0))/x['cfacshr']))
                 .get(['permno','month','divamt']))
nyse_divs_raw
## 1643298
```

```python
nyse_divs = nyse_divs_raw.groupby('permno').apply(lambda x: x.sort_values('month')).reset_index(drop=True)
```

```python
# 计算过去12个月的股利总额和月份数
nyse_divs['div_trailing'] = nyse_divs.groupby('permno')['divamt'].rolling(12, min_periods=1).sum().reset_index(drop=True)
nyse_divs['mths_trailing'] = nyse_divs.groupby('permno')['month'].rolling(12, min_periods=1).count().reset_index(drop=True)

# 计算未来12个月的股利总额和月份数
nyse_divs['div_forward'] = nyse_divs.groupby('permno')['divamt'].shift(-12).rolling(12, min_periods=1).sum().reset_index(drop=True)
nyse_divs['mths_forward'] = nyse_divs.groupby('permno')['month'].shift(-12).rolling(12, min_periods=1).count().reset_index(drop=True)
```

```python
# 筛选出过去和未来12个月均有数据的记录
## 排除例如某股票在过去十二个月内上市或在随后的十二个月内退市等情况，从而可能使得股利变化比率的计算不太有意义的情况
nyse_divs = nyse_divs[(nyse_divs['mths_trailing'] == 12) & (nyse_divs['mths_forward'] == 12)]
# 选择需要的列
nyse_divs = nyse_divs[['permno', 'month', 'div_trailing', 'div_forward']]
# 重置索引
nyse_divs.reset_index(drop=True)
```

```python
# 计算股利变化比率
split_firm_dividends = (nyse_splits
                        .merge(nyse_divs,
                               how='left',
                               on=['permno','month'])
                        .assign(div_ratio = lambda x: np.where(x['div_trailing']>0, x['div_forward']/x['div_trailing'],np.nan))
                        .dropna(subset=['div_ratio'])
                        .get(['permno','month','exdt','div_ratio']))
split_firm_dividends
## 与R结果相同
```

Fama et al. (1969)根据“纽约证券交易所所有证券在相关时间段支付的平均股利”来衡量股利变化。

```python
## 通过计算每个月所有股票的 div_trailing 和 div_forward 值的平均值来计算“市场”股利变化比率（mkt_div_ratio）。
div_mkt = (nyse_divs
           .groupby('month',group_keys=False)
           .agg(div_trailing=('div_trailing', lambda x: np.mean(x.dropna())),
               div_forward=('div_forward', lambda x: np.mean(x.dropna())))
           .reset_index()
           .assign(mkt_div_ratio = lambda x: np.where(x['div_trailing']>0, x['div_forward']/x['div_trailing'],np.nan)))
## 1154
div_mkt
```

考虑以下三种替代理论：

理论1：公司喜欢保持其股价在某一范围内。当股价上涨到一定阈值以上时，公司可能会启动拆分，并提前几周公布拆分生效日期。公司不利用拆分来传递有关公司前景的私人信息。拆分前的股价可能受到的信息驱动，暗示着即将增加股息。

理论2：公司利用拆分来传递有关公司前景的私人信息。公司将在拆分生效日期前几周公布拆分。公司还可能利用股息变化来传递有关公司前景的私人信息。

理论3：公司利用拆分来传递有关未来股息增加的私人信息，这进而传递了有关公司前景的私人信息。公司将在拆分生效日期前几周公布拆分。

提出一些测试，以区分这三个假设中哪个最能解释观察到的现象。哪个理论最符合 FFJR 提出的理论？理论需要如何修改才能更好地符合 FFJR 提出的理论？

1. 测试：
    1. 事件研究分析： 对股票拆分公告和生效日期进行事件研究分析。比较拆分公告前后公司股价的表现，查看是否存在异常收益。如果理论1正确，我们预期会看到股价在拆分公告前上涨，这是因为即将增加股息。如果理论2或3正确，我们预期会看到股价对公司前景的私人信息做出反应
    2. 股息分析： 分析拆分公告与股息变化的关系。如果理论2或3正确，我们预期会看到股息变化与拆分公告同时发生，因为公司可能会利用股息来传递有关其前景的私人信息
    3. 与市场指数比较： 比较宣布拆分的公司与未宣布拆分但股息或股价发生类似变化的公司的行为。这可以帮助孤立出拆分对股价和股息的影响，并区分三种假设中提出的信号理论
    4. 回归分析： 进行回归分析，评估拆分和股息变化对公司绩效指标（如收益、营收和盈利能力）的影响。通过控制其他变量，我们可以确定拆分和股息变化是否携带了公司前景的独特信息，超越了已知信息
    
2. FFJR提出的理论似乎更接近于理论1：公司不利用拆分来传递有关公司前景的私人信息。相反，他们提出，一旦适当考虑了相关股息的信息效应，拆分本身对普通股回报没有净影响
3. 为了更好地与FFJR的理论相符合，可以修改理论1，明确表达公司可能主要是为了调整其股价在一定范围内，而不一定是为了传递有关公司前景的私人信息。这种修改将强调拆分在管理股价水平方面的作用，并且可以更好地解释FFJR所建议的观察到的现象。
