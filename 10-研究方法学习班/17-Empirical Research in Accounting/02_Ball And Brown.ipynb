### 8.On p.418 of Ball and Brown (2019), the authors say, “in this replication we address two issues with the BB68 significance tests.” Do you understand the points being made here?
**第一个问题：更长时间的显著性测试**。BB68 的原始研究主要关注短期内（如公告月或特定几个月内）会计收益信息对股票价格影响的显著性测试。在2019年的复制研究中，通过对前事件（pre-event）、事件（event）、以及后事件（post-event）三个不同时间段的显著性进行测试，而不仅仅是集中在短期内的特定时间点。这种方法可以更全面地评估会计收益信息对股票价格影响的持续时间和强度。

**第二个问题：使用非参数重新采样方法测试显著性**。在 BB68 的原始研究中，他们使用的显著性测试可能对于检测长期趋势不够敏感，且可能依赖于特定的统计模型假设。Ball 和 Brown (2019) 采用了一种非参数的重采样方法（resampling method）来测试会计收益信息与股票价格变化之间关系的显著性。这种方法不依赖于特定的统计分布假设，而是通过从原始数据中随机抽取样本（重采样）来构建经验分布，从而评估观察到的关系在随机情况下出现的概率。这种方法增强了显著性测试的稳健性，使得对结果的解释更加可靠。

### 9.Also, on p.418, Ball and Brown (2019) say “the persistence of PEAD over time is evidence it does not constitute market inefficiency.” What do you make of this argument?
盈余公告后漂移（PEAD）现象的持续存在表明它不构成市场效率的证据。这个论点的核心在于解释市场对会计信息反应的本质及其与市场效率理论的关系。

1. **市场效率理论**：根据有效市场假说（EMH），市场价格在任何给定时间点都充分反映了所有可用信息。如果市场是完全有效的，那么任何公开可获得的信息（包括会计收益信息）都应该立即被市场价格所吸收，从而不会出现基于已公布信息的可预测性价格变动。

2. **PEAD现象**：然而，PEAD现象显示，在收益信息公布后，股票价格往往会在一段时间内继续沿着公告前的趋势漂移，这表明市场没有立即完全吸收新的收益信息。按照严格的市场效率理论，这种延迟反应可能被视为市场效率的缺失。

3. **PEAD与市场效率的关系**：Ball 和 Brown (2019) 提出，PEAD的持续存在其实反映了市场在适应和处理信息方面的复杂性。市场参与者需要时间来解析、验证并对新信息作出反应，特别是在面对复杂的会计信息时。因此，PEAD可能更多地反映了市场参与者处理信息的实际限制，而不是市场效率本身的缺失。

4. **市场效率的动态理解**：他们的观点暗示了对市场效率的更动态和实证的理解。在这个视角下，市场效率不是一个绝对的状态，而是一个过程，市场通过不断的信息处理和价格调整，逐渐反映出信息的价值。PEAD的存在表明市场在不断地接近（但可能永远不会完全达到）充分反映所有信息的理想状态。

Ball 和 Brown (2019) 认为：PEAD的持续存在并不直接指向市场效率的缺失，而是揭示了市场在处理复杂信息时的现实挑战。这种观点强调了在评估市场效率时，需要考虑市场参与者的行为和信息处理机制，以及这些因素如何影响市场价格对信息的反应速度和效率。

### 10.What is the minimum amount of information that our hypothetical genie needs to provide to enable formation of the portfolios underlying TI, NI, and II? What are the rules for construction of each of these portfolios?
在 Ball 和 Brown (1968) 的研究中，假想的genie提供了一个简化的机制来说明如何基于有限的信息构建投资组合，特别是与他们定义的三种投资组合：TI（总信息投资组合），NI（净信息投资组合），以及 II（收益信息投资组合，后文中被 Ball 和 Brown (2019) 称为 AI，即会计信息投资组合）相关的组合。为了能够构建这些投资组合，genie需要提供的最小信息量以及每种组合的构建规则如下：

**1. 最小信息量：**

- 对于所有三种投资组合，genie需要提供的最小信息是每家公司未来一年的收益变化方向（增加或减少）。这种信息足以使投资者区分“好消息”和“坏消息”，并据此决定是买入还是卖出某只股票。

**2. 投资组合的构建规则：**

- **TI（总信息投资组合）**：这个投资组合假设投资者拥有完美的市场前瞻性，能够预知所有信息，包括但不限于会计收益信息。因此，genie需要提供关于所有可能影响股票价格的信息的完整预测。基于这些信息，投资者将购买预期收益增加（好消息）的股票，并卖空预期收益减少（坏消息）的股票。

- **NI（净信息投资组合）**：构建这个投资组合的规则与 TI 类似，但区别在于投资者只对未来股票价格感兴趣，而不是特定类型的信息。genie提供的信息侧重于股票的整体表现预测，而非特定的收益变化或其他详细信息。

- **II/AI（收益信息投资组合）**：对于这个投资组合，genie仅需提供关于公司未来会计收益变化方向的信息，即是否存在“好消息”或“坏消息”。投资者将基于这些简化的收益预测来构建他们的投资组合，买入预期有正面收益变化的股票，卖空预期有负面收益变化的股票。

### 11.Ball and Brown (1968) observe a ratio of NI to TI of about 0.23. What do we expect this ratio to be? Does this ratio depend on the information content of accounting information?
这个比率反映了会计收益信息（作为总信息一部分）在解释股票价格变动中的相对重要性。

**对这个比率的预期：**

这个比率的预期值取决于会计信息在所有影响股票价格变动的信息中所占的比重。如果会计信息内容丰富，能够提供关于公司价值的重要洞察，那么我们可以预期这个比率相对较高，因为会计收益变化将在股票价格变动中占有较大的解释权重。相反，如果会计信息相对较少地反映了公司的真实经营状况或市场信息，这个比率可能会较低。

**这个比率是否依赖于会计信息的信息内容：**

这个比率依赖于会计信息的信息内容。会计信息的信息内容越多，其对股票价格的影响就越大，因此 NI 与 TI 的比率也就越高。这个比率实际上反映了会计收益信息在所有市场信息中所占的比重以及其相对重要性。

- **NI（净信息）**：指的是通过会计收益信息得到的对股票价格有影响的信息量。
- **TI（总信息）**：则包括所有能影响股票价格的信息，不仅仅是会计信息。

如果会计信息能够及时、准确地反映公司的经营状况和前景，则该信息的信息内容多，对投资者决策的影响也就更大，从而在股票价格的形成过程中发挥更重要的作用。因此，会计信息的信息内容越高，NI/TI 的比率也就越高，表明会计信息在所有市场信息中占有更大的比重。

### 12.Consider the paragraph in Ball and Brown (2019, p. 418) beginning “an innovation in BB68 was to estimate …”. How do the discussions of these results differ between Ball and Brown (1968) and Ball and Brown (2019)?
创新点：估算了基于完美预见性的投资策略的收益，特别是区分了总信息（TI）和净信息（NI）的概念，以及会计信息（AI）在内的特定信息类型对投资回报的贡献。提出了两种场景下的假想投资策略：一种是投资者拥有对所有未来信息的完美预见（包括非会计信息），另一种是投资者仅通过会计收益信息进行投资决策。通过比较这两种策略的投资回报，BB68 旨在评估会计信息在所有市场信息中所占的比重。

**Ball 和 Brown (1968) 与 Ball 和 Brown (2019) 讨论结果的不同点：**

1. **分析深度和背景**：Ball 和 Brown (1968) 的原始分析更多地集中在展示会计收益信息对股票价格有影响力的证据上，而 Ball 和 Brown (2019) 则提供了更深入的背景分析，反思了这一发现随时间的持续性和在不同市场环境下的普适性。2019年的文章在过去五十年的文献和实证结果基础上，进一步讨论了这些发现对会计和金融研究的长远影响。

2. **方法论的反思**：Ball 和 Brown (2019) 回顾了 BB68 中的方法论，并讨论了其在后续研究中的应用和发展。他们强调了将会计信息与市场效率相结合的研究方法的重要性，以及这种方法对理解市场如何处理会计信息的影响。

3. **市场效率的理解**：Ball 和 Brown (2019) 对 BB68 的结果提供了更加复杂的市场效率理解。他们指出，尽管会计信息对股票价格有显著影响，但市场并非总是即时或完全有效地反映这些信息，这反映了市场参与者处理和反应信息的限制。他们认为，PEAD 的持续存在并不直接指向市场效率的缺失，而是揭示了市场在处理复杂信息时的状况。

4. **数据和时间的扩展**：Ball 和 Brown (2019) 在讨论他们的结果时，还考虑了更广泛的时间范围和更多国家的数据，这些数据支持了 BB68 发现的普适性。与 BB68 的原始分析相比，后者主要关注特定时间段内的美国市场。

### 13.Consider column (4) of Table 2 of Ball and Brown (2019). Is an equivalent set of numbers reported in BB68? What is the underlying investment strategy associated with this column (this need not be feasible in practice)?
Ball 和 Brown (2019) 文章中表2的第4列显示的是与特定投资策略相关的数字，这种策略假设投资者能够根据会计信息实现的超额回报。具体来说，这列数据反映了如果投资者能够完美地预见哪些公司将公布超过预期的盈利（即“好消息”），以及哪些公司的盈利将低于预期（即“坏消息”），并据此在会计信息公布前买入或卖空这些公司股票，那么他们理论上能够获得的超额回报。

这种投资策略基于的是一种理想化的前提，即投资者可以完美预知未来的会计信息（特别是相对于市场预期的盈利情况），并且能够在没有任何成本或限制的情况下根据这些信息进行交易。实际上，这种投资策略并不可行，因为它要求投资者具有不可能拥有的完美预见能力，并且忽略了交易成本、市场摩擦和信息获取的困难。

在 Ball 和 Brown (1968) 的原始研究中，虽然他们没有直接说明与 Ball 和 Brown (2019) 文章中表2的第4列完全相同的一组数字，但他们确实探索了类似的概念，即通过分析会计信息（特别是盈利报告）对股票价格的影响来评估会计信息的价值。BB68 的研究着眼于会计收益信息如何影响股票回报，并尝试量化这种影响，这与 Ball 和 Brown (2019) 中进一步细化和量化投资策略相关回报的尝试是一致的。

### 14.Heading 6.3 of Ball and Brown (2019) is “Does ‘useful’ disprove ‘meaningless’?” Do you think that “not meaningless” implies “not useless”? Which questions (or facts) does BB68 address in these terms?
“不无意义”（not meaningless）并不直接等同于“不无用”（not useless）。一个信息可能不是完全无意义的，即它可能包含一定程度的相关信息或能在某种程度上反映实际情况，但这并不意味着这个信息就一定是有用的，即能够对投资决策产生实质性的、积极的影响。一个信息要被视为有用，它需要能够提供足够的、及时的、可行动的见解，以帮助市场参与者做出更好的投资决策。

Ball 和 Brown (1968) 在这些术语下主要关注了几个问题（或事实）：

1. **会计收益信息的相关性**：探讨了会计收益信息是否为股票市场参与者提供了有价值的、相关的信息。通过分析股票价格对会计收益公告的反应来评估这一点，从而试图证明会计收益信息并非“无意义”。

2. **信息对股票价格的影响**：进一步分析了会计收益信息在公布前后对股票价格变动的影响，尤其是对意外盈余的反应，来评估这些信息的有用性。通过证明会计收益信息能够影响股票价格，BB68 暗示这些信息是对市场参与者有实际价值的。

3. **市场效率的考量**：通过评估盈余公告后漂移（PEAD）现象，BB68 探讨了市场在吸收和反映会计收益信息方面的效率。虽然 PEAD 现象可能表明市场对会计信息的反应不完全即时或有效，但这并不意味着会计信息是无意义或无用的。相反，PEAD 的存在表明会计收益信息对市场参与者来说是有用的，因为它们能够基于这些信息来调整投资策略。

### 1.2.3 公告日与回报率数据

```python
annc_events = fundq[(fundq['indfmt'] == "INDL") & 
                    (fundq['datafmt'] == "STD") & 
                    (fundq['consol'] == "C") & 
                    (fundq['popsrc'] == "D") &
                    (fundq['fqtr'] == 4) & 
                    (fundq['fyr'] == 12) & 
                    (~fundq['rdq'].isna())]

# 选择列
annc_events = annc_events[['gvkey', 'datadate', 'rdq']]

# 先转换到当月第一天，再向下取整到月份的开始（公告月份）
annc_events['annc_month'] = annc_events['rdq'].dt.to_period('M').dt.to_timestamp('M') - MonthBegin(1)
```

```python
crsp_dates = msi[['date']].copy()
crsp_dates = crsp_dates.sort_values(by='date')

# 添加一个新列td，表示行号，交易日期
crsp_dates['td'] = range(1, len(crsp_dates) + 1)

# 添加一个新列month，该列是date列日期向下取到月初的日期，获取潜在的公告月份
crsp_dates['month'] = crsp_dates['date'].dt.to_period('M').dt.start_time

# 前10行
crsp_dates_head = crsp_dates.head(10)
```

```python
annc_months = crsp_dates[['month', 'td']].copy()
annc_months.rename(columns={'month': 'annc_month', 'td': 'annc_td'}, inplace=True)

# 添加start_td和end_td列，为计算每个收益公告日期t的t-11到t+6月份的回报
annc_months['start_td'] = annc_months['annc_td'] - 11
annc_months['end_td'] = annc_months['annc_td'] + 6

# 前10行
annc_months_head = annc_months.head(10)
```

```python
# 执行交叉合并（cross join），添加一个临时键用于合并
crsp_dates['_tmpkey'] = 1
annc_months['_tmpkey'] = 1
td_link = pd.merge(crsp_dates, annc_months, on='_tmpkey').drop('_tmpkey', axis=1)

# 筛选出满足条件的行，即 td 在 start_td 和 end_td 之间
td_link = td_link[(td_link['td'] >= td_link['start_td']) & (td_link['td'] <= td_link['end_td'])]
td_link['rel_td'] = td_link['td'] - td_link['annc_td'] # 相对交易日
td_link = td_link[['annc_month', 'rel_td', 'date']]

# 筛选annc_month列值为"2001-04-01"的行
td_link[td_link['annc_month']=='2001-04-01'].reset_index(drop=True)
```

```python
# 连接Compustat收益公告日期和CRSP的回报
ccm_link = ccmxpf_lnkhist[
    ccmxpf_lnkhist['linktype'].isin(['LC', 'LU', 'LS']) &
    ccmxpf_lnkhist['linkprim'].isin(['C', 'P'])
].copy()
ccm_link.rename(columns={'lpermno': 'permno'}, inplace=True)

# 计算新列linkenddt的值，如果linkenddt是NA，则用最大的非NA linkenddt值替换
ccm_link['linkenddt'] = ccmxpf_lnkhist['linkenddt'].fillna(ccmxpf_lnkhist['linkenddt'].max(skipna=True))


ccm_link = ccm_link[['gvkey', 'permno', 'linkdt', 'linkenddt']]
ccm_link
```

```python
# 进行内连接
rets_all = pd.merge(annc_events, td_link, on='annc_month')
rets_all = pd.merge(rets_all, ccm_link, on='gvkey')
rets_all = rets_all[
    (rets_all['annc_month'] >= rets_all['linkdt']) &
    (rets_all['annc_month'] <= rets_all['linkenddt']) | rets_all['linkenddt'].isna()
]
rets_all = pd.merge(rets_all, msf, on=['permno', 'date'])
rets_all = pd.merge(rets_all, stocknames, on='permno')

# Nichols 和 Wahlen (2004) 关注的是在纽约证券交易所（NYSE）、美国证券交易所（AMEX）和纳斯达克（NASDAQ）上市的公司
# exchcd 值为 1、2 和 3 
rets_all = rets_all[
    (rets_all['date'] >= rets_all['namedt']) &
    (rets_all['date'] <= rets_all['nameenddt']) &
    (rets_all['exchcd'].isin([1, 2, 3]))
]

rets_all = rets_all[['gvkey', 'datadate', 'rel_td', 'permno', 'date', 'ret']]

# 根据datadate的年份进行过滤，选择1987-2002年的财政年度数据
rets_all['datadate_year'] = rets_all['datadate'].dt.year  # 提取年份
rets_all = rets_all[
    (rets_all['datadate_year'] >= 1987) & 
    (rets_all['datadate_year'] <= 2002)
]

rets_all.drop(columns=['datadate_year'], inplace=True)

rets_all = rets_all.reset_index(drop=True)
rets_all
```

```python
# 选择在t-11到t+6窗口内每个月都有回报的公司
# 分组并计算每组的大小
group_size = rets_all.groupby(['gvkey', 'datadate']).size()
group_size_reset = group_size.reset_index(name='n_obs')

# 筛选出观察数等于最大观察数的记录
max_n_obs = group_size_reset['n_obs'].max()
filtered = group_size_reset.query('n_obs == @max_n_obs')

full_panel = filtered[['gvkey', 'datadate']]
full_panel
```

```python
def read_data(start, end):
    os.environ['VROOM_CONNECTION_SIZE'] = '500000' # 设置环境变量
    df = pd.read_csv('Portfolios_Formed_on_ME.csv',
                    skiprows=start, nrows=end-start, na_values=['-99.99']) # 文本中NA为-99.99
    df.columns = ['date'] + df.columns[1:len(df.columns)].tolist()
    
    # 将日期格式从yyyymm转化为yyyy-mm-01，称为month
    df['month'] = pd.to_datetime(df['date'], format='%Y%m')
    df = df.drop(columns=['date'])
    
    # 宽格式转化为长格式，将不同的回报率分位数转化为单一的列，便于分析
    df = pd.melt(df, id_vars=['month'], var_name='quantile', value_name='ret')
    
    # 计算分位数
    df['ret'] = df['ret']/100
    df['decile'] = np.nan
    df.loc[df['quantile'] == 'Hi 10', 'decile'] = 10 
    df.loc[df['quantile'] == 'Lo 10', 'decile'] = 1
    
    # 对于以 'Dec' 开头的记录，将 'Dec' 后的数字转换为整数并分配给 'decile'
    # 将文本标签转换为易于分析的数值分位数。
    df.loc[df['quantile'].str.startswith('Dec'), 'decile'] = df.loc[df['quantile'].str.startswith('Dec'), 'quantile'].str.replace('Dec', '').astype(int)
    
    df = df.dropna(subset=['decile'])
    df = df.drop(columns=['quantile'])
    
    return df
```

### 1.2.5 业绩新闻变量

Ball 和 Brown (1968) 在处理收益新闻时需要解决的主要问题是一个至今在会计研究中仍然存在的问题：如何衡量收入的非预期成分？或者，如何估计“市场”的收益预期？

Ball 和 Brown (1968) 使用了两种预期收益的度量方法。第一种是一个简单的模型，仅仅“预测今年的收入将与去年相同”（1968年，第163页）。第二种使用了一个模型，估计了公司收入变化与市场收入变化之间的关系，然后将该关系应用于市场收入的观察值。注意，这是一个有趣的变量：股票市场在年初并不知道所有公司的收入。因此，预期是相对于一个相当特殊的信息集而条件化的。实际上，问题在于，给定有关市场收益和市场回报的信息，会计收益的信息是否有助于预测收益的非预期部分。无论如何，主要结果（见著名的图1）对于预期模型的选择是稳健的。

```python
funda_mod = funda[(funda['indfmt']=='INDL')&
                 (funda['datafmt']=='STD')&
                 (funda['consol']=='C')&
                 (funda['popsrc']=='D')]

# 筛选 fyr == 12
news = funda_mod.query('fyr == 12')

news.sort_values(by=['gvkey', 'datadate'], inplace=True)

# 计算滞后值
news['lag_ibc'] = news.groupby('gvkey', group_keys=True)['ibc'].shift()
news['lag_oancf'] = news.groupby('gvkey', group_keys=True)['oancf'].shift()
news['lag_at'] = news.groupby('gvkey', group_keys=True)['at'].shift()
news['lag_fyear'] = news.groupby('gvkey', group_keys=True)['fyear'].shift()

# 筛选年份在1987到2002之间，且满足连续年份条件
news = news.query('1987 <= fyear <= 2002 and lag_fyear + 1 == fyear')

# 计算变化率和增长指标
news['earn_chg'] = np.where(news['lag_at'] > 0, (news['ibc'] - news['lag_ibc']) / news['lag_at'], np.nan)
news['cfo_chg'] = np.where(news['lag_at'] > 0, (news['oancf'] - news['lag_oancf']) / news['lag_at'], np.nan)
news['earn_gn'] = news['earn_chg'] > 0
news['cfo_gn'] = news['cfo_chg'] > 0

# 删除包含 NaN 的行
news.dropna(subset=['cfo_gn', 'earn_gn'], inplace=True)

news = news[['gvkey', 'datadate', 'earn_chg', 'cfo_chg', 'earn_gn', 'cfo_gn']]

def safe_qcut(series, q, **kwargs):
    # 检查唯一值的数量是否足够进行qcut操作
    unique_vals = series.nunique()
    if unique_vals >= q:  # 如果唯一值数量足够，执行qcut
        return pd.qcut(series, q, **kwargs)
    else:  # 如果不够，返回一个全是NaN的序列，或者根据需要进行其他操作
        return pd.Series([np.nan] * len(series), index=series.index)

def assign_deciles(group):
    group['earn_decile'] = safe_qcut(group['earn_chg'], 10, labels=False, duplicates='drop') + 1
    group['cfo_decile'] = safe_qcut(group['cfo_chg'], 10, labels=False, duplicates='drop') + 1
    return group

news = news.groupby('datadate').apply(assign_deciles).reset_index(drop=True)

# 在进行了安全的qcut操作后，有可能生成了含有NaN的行，删除这些行
news.dropna(subset=['earn_decile', 'cfo_decile'], inplace=True)

news
```

Nichols and Wahlen也采用了（t-11，t+6）的时间窗口，具体为公司宣布年度盈利公告前的11个月（t-11）到盈利公告后的6个月（t+6），集中研究在指定窗口内每个月都有回报数据的公司。

3.Nichols 和 Wahlen (2004) 的表格2将累积异常回报定义为“累积原始回报减去公司开始时所属的规模分位数投资组合的累积回报”。除了使用规模分位数投资组合而不是其他市场指数外，这种测量方式与 Ball 和 Brown (1968) 在第168页定义的异常表现指数（API）有何不同？调整在图1复现中描述的测量方法，使其更接近 Ball 和 Brown (1968) 使用的 API 定义（但保留规模分位数作为基准）。这种调整是否显著影响结果？哪种方法更合适？是 Nichols 和 Wahlen (2004) 的方法，还是 Ball 和 Brown (1968) 的方法？

Nichols and Wahlen（2004）先计算公司的累积原始回报率，再减去公司所属的规模十分位组合累计回报率，先累积再相减；而Ball and Brown（1968）先将公司回报率减去市场指数回报率计算出异常回报率，再将异常回报率累积，先相减再累积。
在Nichols and Wahlen（2004）的方法和Ball and Brown（1968）的方法之间的选择取决于研究问题和具体背景。
Ball and Brown（1968）（先相减再累积）适用于以下情况：
   - 当需要以逐步比较的方式分析股票的相对表现时，比如比较一段时间内某只股票相对于市场指数的表现。
   - 当希望将市场指数作为基准来评估单只股票的表现，并且想要了解每个时点的股票相对于基准的表现差异。
Nichols and Wahlen（2004）（先累积再相减）适用于以下情况：
   - 当需要在一个累积时间段内对股票和市场指数的累积表现进行比较时，比如比较一段时间内股票相对于市场指数的累积收益率。
   - 当希望了解整个累积期间内股票的相对表现，而不仅仅是单个时间点的比较。
总的来说，选择哪种方法取决于研究的具体需求，以及希望对股票表现进行逐步比较还是在整个累积时间段内进行综合比较。

```python
plot_data_ball_brown = (merged
                        .dropna(subset=['ret'])
                        .sort_values(by=['gvkey', 'datadate','rel_td'])
                        .reset_index(drop=True)
                        .assign(ret = lambda x: x['ret'] + 1,
                                ew_ret = lambda x: x['ew_ret'] + 1,
                                vw_ret = lambda x: x['vw_ret'] + 1)
                        .assign(abnormal_ret = lambda x: x['ret'] - x['ew_ret'] + 1) # 先相减
                        .assign(abnormal_ret = lambda x: x.groupby(['gvkey', 'datadate'])['abnormal_ret'].cumprod())
                        .groupby(['rel_td', 'earn_gn'])['abnormal_ret']
                        .mean()
                        .reset_index()
                        .assign(Earnings_news = lambda x: x['earn_gn'].map({True: "Good", False: "Bad"}))
                       )
```

```python
# 算法一
# 使用ew_ret直接对ret进行调整
AI = (merged
      .dropna(subset=['ret'])
      .sort_values(by=['gvkey', 'datadate','rel_td'])
      .reset_index(drop=True)
      .assign(adj_ret = lambda x: x['ret'] - x['ew_ret'] + 1)
      .assign(adj_ret = lambda x: x.groupby(['gvkey', 'datadate'])['adj_ret'].cumprod())
      .query("rel_td==0")
      .groupby(['year','earn_gn'])
      .agg({col: ['mean','count'] for col in ['adj_ret']})
      .reset_index()
      .assign(mean_count = lambda x: np.where(x['earn_gn'],(x[('adj_ret','mean')]-1) * x[('adj_ret','count')],
                                              -(x[('adj_ret','mean')]-1) * x[('adj_ret','count')]),
             count = lambda x: x[('adj_ret','count')])
      .groupby(['year'])
      .agg({col: 'sum' for col in [('mean_count',''),('count','')]})
      .reset_index()
      .assign(AI = lambda x: x[('mean_count','')] / x[('count','')])
      [[('year',''),('AI','')]]
     )

AI.columns =[s1 + ""+ str(s2) for (s1,s2) in AI.columns.tolist()]

NI = (merged
      .dropna(subset=['ret'])
      .sort_values(by=['gvkey', 'datadate','rel_td'])
      .reset_index(drop=True)
      .assign(adj_ret = lambda x: x['ret'] - x['ew_ret'] + 1)
      .assign(adj_ret = lambda x: x.groupby(['gvkey', 'datadate'])['adj_ret'].cumprod())
      .assign(adj_ret = lambda x: abs(x['adj_ret'] - 1))
      .query("rel_td==0")
      .groupby(['year'])
      .agg({col: 'mean' for col in ['adj_ret']})
      .reset_index()
      .rename(columns = {'adj_ret':'NI'})
     )

AI_NI = (AI
         .merge(NI,on=['year'],how='inner')
         .assign(AI_NI = lambda x:x['AI']/x['NI']))

(ggplot(AI_NI, aes(x='year', y='AI_NI')) + geom_line())
```

```python
# 算法二
# 网页方法
AI = (merged
      .dropna(subset=['ret'])
      .sort_values(by=['gvkey', 'datadate','rel_td'])
      .reset_index(drop=True)
      .assign(ret = lambda x: x['ret'] + 1,
             ew_ret = lambda x:x['ew_ret'] + 1)
      .assign(ret = lambda x: x.groupby(['gvkey', 'datadate'])['ret'].cumprod(),
             ew_ret = lambda x: x.groupby(['gvkey', 'datadate'])['ew_ret'].cumprod())
      .query("rel_td==0")
      .groupby(['year','earn_gn'])
      .agg({col: ['mean','count'] for col in ['ret','ew_ret']})
      .reset_index()
      .assign(mean_count = lambda x: np.where(x['earn_gn'],(x[('ret','mean')]-x[('ew_ret','mean')]) * x[('ret','count')],
                                              -(x[('ret','mean')]-x[('ew_ret','mean')]) * x[('ret','count')]),
             count = lambda x: x[('ret','count')])
      .groupby(['year'])
      .agg({col: 'sum' for col in [('mean_count',''),('count','')]})
      .reset_index()
      .assign(AI = lambda x: x[('mean_count','')] / x[('count','')])
      [[('year',''),('AI','')]]
     )

AI.columns =[s1 + ""+ str(s2) for (s1,s2) in AI.columns.tolist()]

NI = (merged
      .dropna(subset=['ret'])
      .sort_values(by=['gvkey', 'datadate','rel_td'])
      .reset_index(drop=True)
      .assign(ret = lambda x: x['ret'] + 1,
             ew_ret = lambda x:x['ew_ret'] + 1)
      .assign(ret = lambda x: x.groupby(['gvkey', 'datadate'])['ret'].cumprod(),
             ew_ret = lambda x: x.groupby(['gvkey', 'datadate'])['ew_ret'].cumprod())
      .assign(adj_ret = lambda x: abs(x['ret'] - x['ew_ret']))
      .query("rel_td==0")
      .groupby(['year'])
      .agg({col: 'mean' for col in ['adj_ret']})
      .reset_index()
      .rename(columns = {'adj_ret':'NI'})
     )

AI_NI = (AI
         .merge(NI,on=['year'],how='inner')
         .assign(AI_NI = lambda x:x['AI']/x['NI']))
      
(ggplot(AI_NI, aes(x='year', y='AI_NI')) + geom_line())
```
